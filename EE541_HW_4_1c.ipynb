```python
# Importing the necessary libraries. 
import h5py
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Extracting the data from the file.
open_file = 'mnist_traindata.hdf5'
with h5py.File(open_file, 'r') as columns:
    file_datacols = np.array(columns)
    print(file_datacols)
```

```python
# Converting the data to an array.
open_file = 'mnist_traindata.hdf5' # Training dataset.
with h5py.File(open_file, 'r') as columns:
    xdata = np.float32(np.array(columns['xdata']))
    ydata = np.float32(np.array(columns['ydata']))
print('Dimensions of xdata: ', xdata.shape)
print('Dimensions of ydata: ', ydata.shape)

open_file = 'mnist_testdata.hdf5' # Testing dataset.
with h5py.File(open_file, 'r') as columns:
    xdatatest = np.float32(np.array(columns['xdata']))
    ydatatest = np.float32(np.array(columns['ydata']))

print("The dimensions of xdataset: ", xdatatest.shape)
```

```python
# Training the model.
# x -> xdata; y -> ydata; w -> weight; b -> bias; bs -> batch size; r -> repetition; lr -> learning rate.
def train(x, y, w, b, bs, r, lr):
    samples, dataset = x.shape
    weights = np.zeros([10, 784])
    bias = np.zeros([10, 1])
    loss = 0.0
    loss_list = []
    accuracy = 0
    accuracy_list = []

    for i in range(samples):
        xi = np.reshape(x[i], (784, 1))
        yi = y[i]
        prediction_10 = np.exp(np.matmul(w, xi) + b).reshape([10,])
        d = np.float64(np.sum(prediction_10, axis = 0))
        frac = prediction_10 / d # Fraction part.

        for j in range(y.shape[1]):
            prediction = frac[j]
            
            if yi[j] > 0.0:
                loss -= np.log(prediction)
                weights_value = (1 - prediction) * xi
                bias_value = 1 - prediction
            else:
                weights_value = -prediction * xi
                bias_value = -prediction
            weights[j] -= weights_value[:, 0]
            bias[j] -= bias_value

        max_pred = np.argmax(prediction_10)
        max_y = np.argmax(yi)
        
        if max_pred == max_y:
            accuracy += 1

        if (i + 1) % r == 0:
            loss_list.append(loss / r)
            accuracy_list.append(accuracy / r)
            loss = 0.0
            accuracy = 0

        if i == 0:
            loss_list.append(loss)
            accuracy_list.append(accuracy)
        
        if (i + 1) % bs == 0:
            weights_sum = weights / bs
            bias_sum = bias / bs
            w -= (lr * weights_sum)
            weights = np.zeros([10, 784])
            b -= bias_sum
            bias = np.zeros([10, 1])

    return loss_list, accuracy_list, w, b

# Testing the model.
# x -> xdata; y -> ydata; w -> weight; b -> bias; bs -> batch size; r -> repetition.
def test(x, y, w, b, bs, r):
    samples, dataset = x.shape
    loss = 0.0
    loss_list = []
    accuracy = 0
    accuracy_list = []

    for i in range(samples):
        xi = np.reshape(x[i], (784, 1))
        yi = y[i]
        prediction_10 = np.exp(np.matmul(w, xi) + b).reshape([10,])
        d = np.float64(np.sum(prediction_10, axis = 0))
        frac = prediction_10 / d # Fraction part.

        for j in range(y.shape[1]):
            prediction = frac[j]
            
            if yi[j] > 0.0:
                loss -= np.log(prediction)

        max_pred = np.argmax(prediction_10)
        max_y = np.argmax(yi)
        
        if max_pred == max_y:
            accuracy += 1

        if (i + 1) % r == 0:
            loss_list.append(loss / r)
            accuracy_list.append(accuracy / r)
            loss = 0.0
            accuracy = 0
            
    return loss_list, accuracy_list
```

i. Implement SGD with mini-batch size of 1 (i.e., compute the gradient and update weights after each sample). Record the log-loss and accuracy of the training set and test set every 5,000 samples. Plot the sampled log-loss and accuracy values on the same (respective) figures against the batch number. Your plots should start at iteration 0 (i.e., include initial
log-loss and accuracy). Your curves should show performance comparable to batch gradient descent. How many iterations did it take to acheive comparable performance with batch gradient descent? How does this number depend on the learning rate? (or learning rate decay schedule if you have a non-constant learning rate).

```python
batch = 1
repetition = 5000
learning_rate = 0.01
w = np.zeros([10, 784]) # Resetting the weight values before the fucntion call.
b = np.zeros([10, 1]) # Resetting the bias values before the fucntion call.
train_loss = []
train_accuracy = []
test_loss = []
test_accuracy = []

for i in range(4):
    loss_sum, accuracy, w, b = train(xdata, ydata, w, b, batch, repetition, learning_rate) # Function (train) call.
    train_loss.append(loss_sum)
    train_accuracy.append(accuracy)
    
    total_loss, total_accuracy = test(xdatatest, ydatatest, w, b, batch, repetition) # Function (test) call.
    test_accuracy.append(total_accuracy)
    test_loss.append(total_loss)

train_loss = np.reshape(np.array(train_loss), (52, 1))
train_accuracy = np.reshape(np.array(train_accuracy), (52, 1))
test_loss = np.reshape(test_loss, (8, 1))
test_accuracy = np.reshape(test_accuracy, (8, 1))

# Plotting the log loss graph.
plt.title("Log loss for batch size- 1")
plt.xlabel("5000 samples")
plt.ylabel("Loss in dB")
plt.plot(range(52), 20 * np.log10(train_loss), label = "Loss during the training phase")
plt.plot(range(0, 48, 6), 20 * np.log10(test_loss), label = "Loss during the testing phase")
plt.legend()
plt.show()
```

```python
# Plotting the accuracy graph.
plt.title("Accuracy for batch size- 1")
plt.xlabel("5000 samples")
plt.ylabel("Accuracy of the dataset of 60,000")
plt.plot(range(52), train_accuracy, label = "Accuracy during the training phase")
plt.plot(range(0, 48, 6), test_accuracy, label = "Accuracy during the testing phase")
plt.legend()
plt.show()
```

```python
print("Final training loss: ", float(20 * np.log10(train_loss[51])), 'dB')
print("Final test loss: ", float(20 * np.log10(test_loss[7])), 'dB')
print("Final training accuracy: ", float(train_accuracy[51]) * 100, '%')
print("Final test accuracy: ", float(test_accuracy[7]) * 100, '%')
```
